```python
import os
import pandas as pd
from google.colab import files

def preprocess(files_list, root_dir, review):
    listofFiles = []
    for (dirpath, dirnames, filenames) in os.walk(root_dir):
        listofFiles += [os.path.join(dirpath, file) for file in filenames]

    labeled_class = []
    reviews = []
    actual_class = []

    for j in listofFiles:
        labeled_class.append(review)
        k = str(open(j, encoding='utf-8').read())
        reviews.append(k)
        actual_class.append(str(j.split('/')[-2].split('_')[0]))

    data = pd.DataFrame({'labeled_class': labeled_class, 'review': reviews, 'actual_class': actual_class})
    return data

# Assuming you have already uploaded the zip files using the previous code
# If not, you can upload them using files.upload()

# Modify the paths accordingly if the structure inside the zip files is different
negative_data = preprocess(negative_list, "negative_reviews", "negative")
positive_data = preprocess(positive_list, "positive_reviews", "positive")

# Example: Display the first few rows of the negative_data DataFrame
print(negative_data.head())
print(positive_data.head())
```

```python
data = pd.concat([positive_df, negative_df], ignore_index=True)

# Label encoding for the target variable
data['target'] = data.apply(lambda row: 1 if (row['labeled_class'] == 'positive' and row['actual_class'] == 'truthful') else 0, axis=1)

# Shuffle the data
data = data.sample(frac=1).reset_index(drop=True)

# Selecting only the 'review' and 'target' columns from the merged DataFrame
data = data[['review', 'target']]
```

```python
def extract_tokens(df, stop_words):
    review_tokenized = []
    lmt = WordNetLemmatizer()

    for index, datapoint in df.iterrows():
        tokenize_words = word_tokenize(datapoint["review"].lower(), language='english')
        pos_word = pos_tag(tokenize_words)

        # Remove stopwords and lemmatize
        tokenize_words = [
            "_".join([lmt.lemmatize(word), pos]) for word, pos in pos_word
            if (word not in stop_words and len(word) > 2)
        ]

        token = ' '.join(tokenize_words)
        review_tokenized.append(token)

    df["review_tokenized"] = review_tokenized
    return df

# Retrieve stopwords once
stop_words = set(stopwords.words("english"))

# Apply the tokenization and lemmatization
data = extract_tokens(data, stop_words)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming you've already fit_transformed your data with TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(data)

# Get feature names
feature_names = np.array(vectorizer.get_feature_names_out())

# Calculate TF-IDF values for each feature
tfidf_values = tfidf_matrix.sum(axis=0).A1

# Sort feature indices based on TF-IDF values
sorted_indices = np.argsort(tfidf_values)[::-1]

# Select top N features
n = 3
top_n_features = feature_names[sorted_indices][:n]

# Print the top N features and their TF-IDF values
for feature, tfidf_value in zip(top_n_features, tfidf_values[sorted_indices][:n]):
    print(f"{feature}: {tfidf_value}")
```

```python
def train_rfc(X,y):
    n_estimators = [100]
    min_samples_split = [2]
    min_samples_leaf = [1]
    bootstrap = [True]
    parameters = {'n_estimators': n_estimators, 'min_samples_leaf': min_samples_leaf,
                  'min_samples_split': min_samples_split}
    clf = GridSearchCV(RandomForestClassifier(verbose=1,n_jobs=-1), cv=4, param_grid=parameters)
    clf.fit(X, y)
    return clf
```
